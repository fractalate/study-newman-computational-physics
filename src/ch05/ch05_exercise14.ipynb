# Chapter 5, Exercise 14

We know that the total force of gravity between two masses is equal to

$$
    F = G \frac{m_1 m_2}{r^2}.
$$

In our case with the point mass and plate, $m_1 = 1$ and $m_2 = \sigma \, \mathrm{d}x \, \mathrm{d}y$

$$
    F = G \frac{\sigma \, \mathrm{d}x \, \mathrm{d}y}{r^2}.
$$

We'll then identify the component of the force in the $z$ direction:

Let $(x, y)$ be a point on the plate, then the distance from the point mass to the plate is $r = \sqrt{x^2 + y^2 + z^2}$. So then the component of the force along the $z$ axis would be $F_z = F \cos(\theta)$. Since we know that $cos(\theta) = z/\sqrt{x^2+y^2+z^2}$ we have

$$
F_z = \frac{F z}{\sqrt{x^2+y^2+z^2}}
$$

Now, using the formula for $F$ we can say that

$$
    F = G \frac{\sigma \, \mathrm{d}x \, \mathrm{d}y}{x^2 + y^2 + z^2}.
$$

Then,

$$
    F_z = G \frac{z \sigma \, \mathrm{d}x \, \mathrm{d}y}{(x^2 + y^2 + z^2)^{3/2}}
$$

Finally, the total force along the $z$ axis could be calulated with the following integral:

$$
    G z \sigma \int_{-\frac{L}{2}}^{\frac{L}{2}} \int_{-\frac{L}{2}}^{\frac{L}{2}} \frac{\mathrm{d}x \, \mathrm{d}y}{(x^2 + y^2 + z^2)^{3/2}}
$$
